#### Education number is ascending on the level of completed education, so we can delete the strings column. Also drop 

```python
data[['education-num','education']].groupby(['education-num','education']).size()
```

```python
data.drop('education', axis = 1, inplace= True)
```

```python
data[(data['capital-loss'] != 0) & (data['capital-gain'] == 0)]
```

```python
data['capital-gain-loss'] = data['capital-gain'] - data['capital-loss']
```

```python
data.drop(['capital-gain','capital-loss'], axis = 1, inplace = True)
```

```python
data['marital-status'].unique()
```

```python
data1 = data[['age','workclass','fnlwgt','education-num','occupation','race','sex','hours-per-week','class']]
```

```python
data1.head()
```

#### Now let's make some visuals!

```python
sns.distplot(a=data['age'])
```

```python
data['age'].hist()
```

```python
g=sns.FacetGrid(data=data1, col='sex')
g.map(sns.distplot, 'age')
```

```python
sns.pairplot(data1[['age','education-num','hours-per-week','sex']], hue = 'sex')
```

```python
fig = plt.figure(figsize=(10,7))
sns.distplot(data1[data1['sex']==' Male']['age'], hist=False, color='blue', label='Male')
sns.distplot(data1[data1['sex']==' Female']['age'], hist=False, color='pink', label='Female')

plt.title('Male vs Female')
plt.legend(bbox_to_anchor = (1,0.5))
plt.show()
```

```python
fig = plt.figure(figsize = (10,7))
sns.countplot(x=data1['age'], hue = data1['sex'])
plt.legend(bbox_to_anchor = (1,0.2))
plt.xticks(np.arange(0, data1['age'].max()-data1['age'].min(), step=9))
plt.show()
```

```python
fig = plt.figure(figsize = (10,7))
sns.countplot(x=data1['age'], hue = data1['class'])
plt.legend(bbox_to_anchor = (.94,0.5))
```

```python
fig = plt.figure(figsize=(10,7))
sns.distplot(data1[data1['class']==' <=50K']['age'], hist=False, color='blue', label='<=50K')
sns.distplot(data1[data1['class']==' >50K']['age'], hist=False, color='pink', label='>50K')

plt.title('Classes distribution')
plt.legend(bbox_to_anchor = (1,0.5))
plt.show()
```
